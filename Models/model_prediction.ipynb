```python
import joblib
import pandas as pd
```

```python
df = pd.read_csv('test.csv')
ids = df['ID']            
X = df.drop(columns=['ID']).values
model = joblib.load('models/LR_cluster.pkl')
y_pred = model.predict(X)
result = pd.DataFrame({
    'ID': ids,
    'y_pred': y_pred
})
result.to_csv('prediction.csv', index=False)
```
